## Outlier Treatment

```python
for col in df.columns:

  if df[col].dtype !='object':
    sns.boxplot(df[col])
    plt.show()
    print('---'*50)
```

```python
df.columns
```

```python
out_col = ['age', 'fnlwgt', 'education-num',  'capital-gain', 'capital-loss', 'hours-per-week']
```

```python
# IQR Techniuqe of Outlier removing

for col in out_col:

  Q1 = df[col].quantile(0.25)
  Q3 = df[col].quantile(0.75)

  IQR = Q3-Q1

  UL = Q3+1.5*IQR
  LL = Q1-1.5*IQR

  df = df[( df[col]>=LL ) & ( df[col] <= UL )]
```

```python
for col in out_col:

  sns.boxplot(df[col])
  plt.show()
  print('---'* 50)
```

## Checking Multicolinearity

```python
# VIF
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
x = df.drop('income', axis=1) # independent
y = df['income']  #dependent
```

```python
vif_data = pd.DataFrame()

vif_data['Columns'] = x.columns
vif_data['VIF'] = [variance_inflation_factor(x.values, i) for i in range(len(x.columns))]

vif_data
```
